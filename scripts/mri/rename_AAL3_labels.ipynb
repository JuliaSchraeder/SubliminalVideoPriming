```python
# Define the replacement function
def replace_labels(label):
    label = label.strip("'")  # Remove leading and trailing quotes if they exist
    hemisphere = ''
    if '_L' in label:
        hemisphere = ' L'
    elif '_R' in label:
        hemisphere = ' R'
    
    core_label = label.replace('_L', '').replace('_R', '')
    abbr, desc = aal3_info.get(core_label, (core_label, core_label))
    return pd.Series([label, abbr + hemisphere, desc + ' ' + hemisphere])

# Apply the replacement to the DataFrame
df[['Label', 'Possible_Abbreviation', 'Anatomical_Description']] = df['Label'].apply(replace_labels)

# Save the updated DataFrame to a new Excel file
updated_file_path = 'W:/Fmri_Forschung/Allerlei/JuliaS/GitHub/SubliminalVideoPriming/data/mri/updated_posthoc_Labels.xlsx'
df.to_excel(updated_file_path, index=False)

df.head()
```
